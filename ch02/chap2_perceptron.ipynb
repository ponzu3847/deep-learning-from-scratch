## まとめ
* パーセプトロンは入出力を備えたアルゴリズム。ある入力を与えたら、決まった値が出力される
* パーセプトロンでは、「重み」と「バイアス」をパラメータとして設定する
* パーセプトロンを用いれば、NANDやANDなどの論理回路を表現できる
* XORゲートは単層のパーセプトロンでは表現できない
* ２層のパーセプトロンを用いればXORゲートを表現できる
* 単層のパーセプトロンでは線形領域だけしか表現できないのに対して、多層のパーセプトロンは非線形領域を表現できる
* 多層のパーセプトロンは、理論上コンピュータを表現できる

```
import numpy as np
```

## パーセプトロンとは

###  ANDゲートの実装

```
def AND(x1,x2):
  w1,w2,theta=0.5,0.5,0.7
  tmp=x1*w1+x2*w2
  if tmp<=theta:
    return 0
  else:
    return 1

print(AND(1,1))
print(AND(1,0))
print(AND(0,1))
print(AND(0,0))
```

###  NANDゲートの実装

```
def NAND(x1,x2):
  w1,w2,theta=-0.5,-0.5,-0.7
  tmp=x1*w1+x2*w2
  if tmp<=theta:
    return 0
  else:
    return 1

print(NAND(1,1))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(0,0))
```

### ANDゲートの実装(重みとバイアスによる実装)

```
def AND(x1,x2):
  x=np.array([x1,x2])
  w=np.array([0.5,0.5])
  b=-0.7
  tmp=np.sum(w*x)+b
  
  if tmp<=0:
    return 0
  else:
    return 1 

print(AND(1,1))
print(AND(1,0))
print(AND(0,1))
print(AND(0,0))
```

### NANDゲートの実装(重みとバイアスによる実装)

```
def NAND(x1,x2):
  x=np.array([x1,x2])
  w=np.array([-0.5,-0.5])
  b=0.7
  tmp=np.sum(w*x)+b
  
  if tmp<=0:
    return 0
  else:
    return 1 

print(NAND(1,1))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(0,0))
```

### ORゲートの実装(重みとバイアスによる実装)

```
def OR(x1,x2):
  x=np.array([x1,x2])
  w=np.array([0.5,0.5])
  b=-0.2
  tmp=np.sum(w*x)+b
  
  if tmp<=0:
    return 0
  else:
    return 1 

print(OR(1,1))
print(OR(1,0))
print(OR(0,1))
print(OR(0,0))
```

### NAND,OR,ANDを組み合わせてXORを実装
* NAND,OR,ANDは線形な領域しか分離できないが、NAND,OR,ANDを組み合わせる（層を重ねる）ことで、XORのような非線形な領域を分離することができるようになる

```
def XOR(x1,x2):
  s1=NAND(x1,x2)
  s2=OR(x1,x2)
  return AND(s1,s2)

print(XOR(1,1))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(0,0))
```
